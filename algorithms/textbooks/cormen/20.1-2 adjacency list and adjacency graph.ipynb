## Questions from Corment et al

Given an adjacency-list representation of a directed graph, how long
does it take to compute the out-degree of every vertex? How long does it take to compute the in-degrees?

To compute the out-degrees for every vertex requires one to visit all nodes calculate the length of the list of adjacent nodes (edges). If the list of edges is a linked list we have to go through the entire list to find the length so that the running time becomes $O(V+E)$ if it's instead an array then the running time $O(V)$

In-degrees means we have to traverse the other nodes and count each time the nodes are pointing to the node of interest. That is we again need $O(V + E)$ running time.
