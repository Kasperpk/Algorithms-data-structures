```python
20.1-5
The square of a directed graph G = (V, E) is the graph G<sup>2</sup> = (V, E<sup>2</sup>)
such that (u, v)
∈ E<sup>2</sup> if and only if G contains a path with at most two
edges between u and v. Describe efficient algorithms for computing G2
from G for both the adjacency-list and adjacency-matrix representations
of G. Analyze the running times of your algorithms.

Stated in plain english what we have to do here is to create a graph (G<sup>2</sup>) which only creates an edge between `u` and `v` if there is no more than two ways of getting from `u` to `v`. Let's take an example of a graph <pre> graph = {
      1: [2, 2, 3],  # two edges to 2, one to 3
      2: [1, 1],
      3: [1, 3]      # 3→3 is a self-loop
  } </pre>

Here there is at most two edges between all the vertices so the new graph will look similar to this one. We have to find a way to count the number of ways of getting from `u` to `v`.


### The adjacency list

To solve this exercise as seen above we run two nested loops because that gives us the edges that are at most 2 paths away. These are added to the new graph which is constructed with a dictionary comprehension. It ten loops over each vertex in the graph, then for each neighbor it adds those edges and then for each neighbors neighbor it adds the edges. The running time here is `O(V*E)`

### The adjacency matrix
```

```python
def to_simple_undirected(graph):
      new_graph = {v: set() for v in graph}
      for vertex in graph:
          for neighbor in graph[vertex]:
              if neighbor == vertex:
                  continue
              new_graph[vertex].add(neighbor)
              new_graph[neighbor].add(vertex)

      return new_graph
```

```python
graph = {
      1: [2, 2, 3],  # two edges to 2, one to 3
      2: [1, 1],
      3: [1, 3]      # 3→3 is a self-loop
  }
```

```python
def undirected_to_directed(graph):
    visited = {v: set() for v in graph}
    for vertex in graph:
        for neighbor in graph[vertex]:
            if neighbor == vertex:
                continue
            visited[vertex].add(neighbor)
            visited[neighbor].add(vertex)
    return visited
```
