  This one is more conceptual than algorithmic. Start with part 1 — what's the expected lookup time for a hash table with chaining under uniform
  independent hashing?

 1. Expected time to look up an edge using a hash table with chaining and uniform independent hashing?

If the lookup is uniform and independent then the expected lookup time is `O(1)` i recollect that hash tables have lookup in that time. More precisely, with chaining the expected time is $O(1 + \alpha)$ where $\alpha$ is the load factor (number of edges / table size). If the table
   is sized proportionally to the number of edges, $\alpha$ is constant and lookup is O(1).

2. What are the disadvantages of hash tables for adjacency lists?

Here are the key disadvantages of hash tables for adjacency lists:

  1. No ordering — a linked list preserves insertion order, a hash table does not. Some graph algorithms depend on iterating neighbors in a
  predictable order.
  2. Space overhead — hash tables have overhead from empty buckets and metadata. A linked list uses exactly as much space as there are edges — no
  waste.
  3. Worst case — hash tables have O(1) expected lookup but O(n) worst case if many keys collide. A sorted structure gives guaranteed bounds.
  4. Iteration is slower — iterating over all neighbors in a hash table requires scanning all buckets including empty ones, whereas a linked list
  iterates in exactly O(degree) time.


  3. Suggest an alternative data structure — does it have disadvantages compared to hash tables?

a balanced BST (like a red-black tree) gives you:

  - O(log degree) edge lookup — better worst case than hash table
  - Ordered iteration — neighbors in sorted order
  - No collision issues — guaranteed bounds, no load factor to manage
  - O(log degree) insert and delete

  Disadvantage compared to hash table?
  Lookup is O(log degree) instead of O(1) expected. So you trade the hash table's average-case speed for guaranteed worst-case performance and
  ordering.
