  $$\sum_k b_{0k} \cdot b_{1k} = (-1)(1) + (0)(-1) = -1$$

  That's -1 because there is an edge between vertex 0 and vertex 1.

  Now pick off-diagonal entry (0,2):

  $$\sum_k b_{0k} \cdot b_{2k} = (-1)(0) + (0)(1) = 0$$

  That's 0 because there is no edge between vertex 0 and vertex 2.

  ---
  So the pattern is:
  - Diagonal $(i,i)$ = degree of vertex i
  - Off-diagonal $(i,j)$ = -1 if there is an edge between i and j, 0 otherwise

  represent a graph as an incidence matrix B, then the product BB^T automatically tells you:

  - How many edges each vertex has (the diagonal)
  - Which vertices are connected (the off-diagonal)

  In other words BB^T is essentially a version of the adjacency matrix combined with a degree matrix — all from a single matrix multiplication.

  Why does this matter?

  It connects graph theory to linear algebra. This is the foundation of a whole field called spectral graph theory, where you analyze graphs by
  studying the eigenvalues and eigenvectors of matrices derived from them. It has real applications in:
  - Google's PageRank algorithm
  - Clustering and community detection in social networks
  - Dimensionality reduction in machine learning
