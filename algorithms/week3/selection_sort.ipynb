The selection sort algorithm is a classic sorting algorithm. It works as follows.

1. Initialize the len of the array as a variable *n*
2. Loop over the range 0-n so that we check every element
3. For every element loop again starting at next element, check if that element is less than the current number in the array.
4. If so then swapping min_index to that number, so here we essentially find the smallest number in the array.
5. When the inner loop finishes then we have found the index of the minimum value, we can then swap the current index with that number.

Here is a sketch of how it works



This sketch alludes to having a nested loop because we can see that we go for each element and compare against all the other elements in the array. This also means that the running time of the algorithm is quadratic that is *O(n^2)*

```python
def selection_sort(array):
    n = len(array)
    for index in range(0, n):
        min_index = index
        for number in range(index+1, n):
            if array[number] < array[min_index]:
                min_index = number
        array[index], array[min_index] = array[min_index], array[index]

    return array
```

```python
array = [5,2,4,3,2,6]
```

```python
selection_sort(array=array)
```

The running time of the merge sort is at worst *O(n log n)* which we can analyze using a recursive tree or a recursive relation.



There can be made mathematical proofs that show that merge sort is asympotically optimal because no algorithm can sort faster an array faster. This is where we use ideas of mathematics and proofs in computer science.

```python
def merge(b,c):
    d = []
    index = 0
    while b or c:
        # if there are elements left in both a and b, then select the first elements in both
        # arrays and compare them to find the smallest one.
        if b and c:
            first_element_b = b[index]
            first_element_c = c[index]
            if first_element_b <= first_element_c:
                d.append(first_element_b)
                b.remove(first_element_b)
            else:
                d.append(first_element_c)
                c.remove(first_element_c)

        # if only we have elements in b then take the element from that array
        elif b and not c:
            d.append(first_element_b)
            b.remove(first_element_b)

        # if only we have elements in c then take the element from that array
        elif c and not b:
            d.append(first_element_c)
            c.remove(first_element_c)
    return d
```
