```python
def largest_product_in_grid(grid):
    import numpy as np 
    up = [grid[i-1][i] for i in range(1,5)]
    down = [grid[i+1][i] for i in range(1,5)]
    left = [grid[i][i-1] for i in range(1,5)]
    right = [grid[i][i+1] for i in range(1,5)]

    up_prod = int(np.cumulative_prod(up)[-1])
    down_prod = int(np.cumulative_prod(down)[-1])
    left_prod = int(np.cumulative_prod(left)[-1])
    right_prod = int(np.cumulative_prod(right)[-1])

    return  max(up_prod, down_prod, left_prod, right_prod)
```
