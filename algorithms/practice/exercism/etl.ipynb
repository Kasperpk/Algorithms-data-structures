```python
letter_points =  {1: ["a", "e", "i", "o", "u", "l", "r", "s", "t"], 
 2: ["d", "g"],
 3: ["b", "c", "m", "p"], 
 4: ["f", "h", "v", "w", "y"], 
 5: ["k"],
 8: ["j", "x"], 
 10: ["q", "z"]
}

def transform(legacy_data):
    letters = []
    for value in legacy_data.values():
        for letter in value:
            letters.append(letter.lower())
    transformed_letter = {}
    for letter in sorted(letters):
        for key, values in letter_points.items():
            if letter in values:
                transformed_letter[letter] = key

    return transformed_letter
```

```python
transform({1: ["A", "E"], 2: ["D", "G"]})
```
