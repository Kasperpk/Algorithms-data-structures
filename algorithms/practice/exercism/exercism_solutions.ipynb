```python
def steps(number):
    '''
    Function to calculate the number of steps it takes to reach 1
    according to the rules of the Collatz Conjecture
    '''
    num_steps = 0
    if number <= 0:
        raise ValueError ('The input number must be positive')
    while number != 1:
        if number%2 == 0:
            number = number / 2
            num_steps += 1
        else:
            number = (number * 3) + 1
            num_steps += 1
    return num_steps
```

A very interesting problem here of trying to return the dart score based on the following:

Calculate the points scored in a single toss of a Darts game.

Darts is a game where players throw darts at a target.

In our particular instance of the game, the target rewards 4 different amounts of points, depending on where the dart lands:

Our dart scoreboard with values from a complete miss to a bullseye

* If the dart lands outside the target, player earns no points (0 points).
* If the dart lands in the outer circle of the target, player earns 1 point.
* If the dart lands in the middle circle of the target, player earns 5 points.
* If the dart lands in the inner circle of the target, player earns 10 points.

The outer circle has a radius of 10 units (this is equivalent to the total radius for the entire target), the middle circle a radius of 5 units, and the inner circle a radius of 1. Of course, they are all centered at the same point — that is, the circles are concentric defined by the coordinates (0, 0).

Given a point in the target (defined by its Cartesian coordinates x and y, where x and y are real), calculate the correct score earned by a dart landing at that point.

```python
def score(x, y):
    '''
    Function that returns the score in a game of dart
    '''
    if (x - 0)**2 + (y - 0)**2 <= 1**2:
        return 10
    elif (x - 0)**2 + (y - 0)**2 <= 5**2:
        return 5
    elif (x - 0)**2 + (y - 0)**2 <= 10**2:
        return 1
    else:
        return 0
```

The insight used here is that we can find out if the coordinate (x,y) is within the different areas of the dart by calculating (x-center)<sup>2</sup> + (y-center)<sup>2</sup> <= radius<sup>2</sup>

```python
def classify(number):
    """ A perfect number equals the sum of its positive divisors.

    :param number: int a positive integer
    :return: str the classification of the input integer
    """
    if number <= 0:
        raise ValueError("Classification is only possible for positive integers.")
    sum = 0

    for num in range(1, int(number/2)+1):
        if number%num == 0:
            sum = sum + num
    if sum == number:
        return 'perfect'
    elif sum >= number:
        return 'abundant'
    else:
        return 'deficient'
```

Given a sorted array of distinct integers and a target value, return the index if the target is found. If not, return the index where it would be if it were inserted in order.

You must write an algorithm with O(log n) runtime complexity.

The solution here below might have some challenges, but right now i am not sure how to overcome them. I wanted to use binary search but that repeatedly shrinks the array and so how do you return the index if not found.

There are several problems with the algorithm. For example if the distance between elements are above 99999 it will not update the closet element index. It also runs in linear time for which i know there are better solutions. 

```python
class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        '''
        Function that returns the index of a target value in array
        of sorted numbers. If not found it returns the index which it
        should be placed in.
        :param nums: Array of sorted numbers
        :param target: Target value to find
        :return: Index of the position of target value in array
        '''

        # Tracking a closest element and index to return if
        # Target not found in array
        closest_element = 99999
        closest_element_index = 0

        # Using a for loop to find the target value or it's
        # Closest element index
        for index in range(0, len(nums)):

            # If the target - current index value in array is
            # less than current closest element then
            # This is a better position for it in the array
            # The difference has to be the closest to 0 and not
            # Large negative values.
            if abs(target - nums[index]) < closest_element:
                closest_element = target - nums[index]
                closest_element_index = index

            # if we find the target we just return the index
            if target == nums[index]:
                return index
        # Otherwise we return the index of the element closest to the
        # Target value
        else:
            if target >= nums[closest_element_index]:
                return closest_element_index + 1
            else:
                return closest_element_index
```

For a more optimal solution i had to talk to Claude about it, because i had tried my way with binary search but didn't know how to return the index of where the target should be placed if not found. It turns out there is quite a brilliant solution that defines a left and right variable in the initialization of the algorithm. Here `left` is 0 and `right` is the len of the array - 1. A while loop then proceeds until left no longer is smaller or equal to right. A variable `mid` is calculated and if nums[mid] is equal to target. Otherwise it checks if target is less than the mid point, if so `right` is changed to mid - 1, so the array is essentially halfed in either direction, the mid point is calculated again and the code proceeds in that way. This is quite a brilliant solution that i don't think i would have thought about, but using the left and right pointers (in general using pointers) just seems like a very crux move.

```python
class Solution:
    def search_insert(self, nums: List[int], target: int) -> int:
        left, right = 0, len(nums) - 1
        while left <= right:
            mid = (left + right) // 2
            if nums[mid] == target:
                return mid
            elif target < nums[mid]:
                    right = mid - 1
            else:
                left = mid + 1

        # When not found, left is the correct insert position
        return left
```

The following problems is one of finding the *k* largest values in a list of numbers. I had several iterations of this algorithm sought out. The first one uses a while loop and iterates over the stream of numbers, simply comparing the value starting at index 1 with index 0, then keeping track of the largest index.

```python
def k_largest(stream, k, ) -> list[int]:
    largest_values = []
    while len(largest_values) != k:
        largest_index = 0
        for index in range(1, len(stream)):
            if stream[index] > stream[largest_index]:
                largest_index = index
        largest_values.append(stream[largest_index])
        stream.remove(stream[largest_index])

    return largest_values
```

```python
import heapq

def k_largest(stream, k) -> list[int]:
    heap = stream[:k]
    heapq.heapify(heap)            # O(k)

    for num in stream[k:]:
        if num > heap[0]:          # heap[0] is always the minimum
            heapq.heapreplace(heap, num)

    return heap
```

```python
def is_pangram(sentence):
    alphabet = ["a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z"]
    for character in sentence:
        if character.lower() in alphabet:
            alphabet.remove(character.lower())
    if alphabet:
        return False
    return True
```
