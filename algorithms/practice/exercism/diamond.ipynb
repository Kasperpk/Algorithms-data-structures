# Diamond Kata

Given a letter, print it in a diamond shape. The diamond starts with `A`, with the
supplied letter at its widest point, then mirrors back down to `A`.

**Example** — input `'C'`:

```
  A
 B B
C   C
 B B
  A
```

## Requirements

- The first and last rows each contain a single `A`.
- Every other row contains exactly two identical letters.
- Each row has as many trailing spaces as leading spaces (possibly zero).
- The diamond is symmetric both horizontally and vertically.
- The diamond is square: width equals height.
- The top half runs in ascending order; the bottom half mirrors it in descending order.
- The four corners are triangles of spaces.

## What I learned

I developed a working solution fairly quickly once I spotted the pattern and the rules
for building the diamond. On the second iteration I improved the algorithm: instead of
running a second loop to build the bottom half, I could just reverse the work already
done in the first loop. That insight didn't occur to me at first, but it's a genuinely
satisfying one. It probably doesn't improve the running time — reversing a list is still
O(n) — but it makes the solution noticeably more compact.

````python
import string
letters = string.ascii_uppercase

def rows(letter):
    diamond = []
    a_index = letters.index(letter)
    if letter == 'A':
           return ["A"]
    
    for char_num in range(0, a_index+1):
        num_space = a_index - char_num
        if letters[char_num] == 'A':
            line = " "*num_space + letters[char_num] + " "*num_space
        else:
            line = " "*num_space + letters[char_num] + " "*(char_num+char_num-1) + letters[char_num] + " "*num_space
        diamond.append(line)

    for char_num in range(a_index-1, -1, -1):
        num_space = a_index - char_num
        if letters[char_num] == 'A':
            line = " "*num_space + letters[char_num] + " "*num_space
        else:
            line = " "*num_space + letters[char_num] + " "*(char_num+char_num-1) + letters[char_num] + " "*num_space
        diamond.append(line)

    return diamond
````

````python
import string
letters = string.ascii_uppercase

def rows(letter):
    diamond = []
    a_index = letters.index(letter)
    if letter == 'A':
           return ["A"]
    
    for char_num in range(0, a_index+1):

        # Num space defines how many spaces is needed between before and after characters
        # That depends on the index of the input letter as well as the current letter
        num_space = a_index - char_num

        # A is handled specifically because it's not added twice  
        if letters[char_num] == 'A':
            line = " "*num_space + letters[char_num] + " "*num_space

        # Every other number's line is defined with open space, character and space, character and 
        # Space again. The rules are as follows
        # The opening and closing space is equal to the num_space
        # The middle space is equal to 2 times the index of the number in the alphabet -1
        else:
            line = " "*num_space + letters[char_num] + " "*(char_num+char_num-1) + letters[char_num] + " "*num_space
        diamond.append(line)

    return diamond + list(reversed(diamond[:-1]))
````
