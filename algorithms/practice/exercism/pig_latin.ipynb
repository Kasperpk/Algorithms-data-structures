# Pig Latin

## Introduction

Your parents have challenged you and your sibling to a game of two-on-two basketball. Confident they'll win, they let you score the first couple of points, but then start taking over the game. Needing a little boost, you start speaking in **Pig Latin**, which is a made-up children's language that's difficult for non-children to understand. This will give you the edge to prevail over your parents!

## Instructions

Your task is to translate text from English to Pig Latin. The translation is defined using four rules, which look at the pattern of vowels and consonants at the beginning of a word.

- **Vowels:** the letters `a`, `e`, `i`, `o`, and `u`
- **Consonants:** the other 21 letters of the English alphabet

---

### Rule 1

If a word begins with a **vowel**, or starts with `"xr"` or `"yt"`, add an `"ay"` sound to the end of the word.

| Input | Output | Reason |
|-------|--------|--------|
| `"apple"` | `"appleay"` | starts with vowel |
| `"xray"` | `"xrayay"` | starts with `"xr"` |
| `"yttria"` | `"yttriaay"` | starts with `"yt"` |

---

### Rule 2

If a word begins with one or more **consonants**, move those consonants to the end of the word, then add an `"ay"` sound.

| Input | Intermediate | Output |
|-------|-------------|--------|
| `"pig"` | `"igp"` | `"igpay"` |
| `"chair"` | `"airch"` | `"airchay"` |
| `"thrush"` | `"ushthr"` | `"ushthray"` |

---

### Rule 3

If a word starts with zero or more consonants followed by `"qu"`, move those consonants and the `"qu"` to the end of the word, then add an `"ay"` sound.

| Input | Intermediate | Output |
|-------|-------------|--------|
| `"quick"` | `"ickqu"` | `"ickquay"` |
| `"square"` | `"aresqu"` | `"aresquay"` |

---

### Rule 4

If a word starts with one or more consonants followed by `"y"`, move the consonants preceding the `"y"` to the end of the word, then add an `"ay"` sound.

| Input | Intermediate | Output |
|-------|-------------|--------|
| `"my"` | `"ym"` | `"ymay"` |
| `"rhythm"` | `"ythmrh"` | `"ythmrhay"` |


```python
def translate(text):
    

    consonants = ['b', 'c', 'd', 'f', 'g', 'h', 'j', 'k', 'l', 'm', 'n', 'p', 'q', 'r', 's', 't', 'v', 'w', 'x', 'z']

    
    words = text.split(' ')
    intermediates = []
    print(words)
    for word in words:

        if word[0] + word[1] == 'yt' or (word[0] not in consonants and word[0] != 'y') or word[0] + word[1] == 'xr':
            intermediate = word + 'ay'
            intermediates.append(intermediate)
            continue

        track_word = 0
        letter = word[track_word]
        consonant_text = '' 
        while letter in consonants or (word[track_word] == 'y' and word[0] == 'y'):
            consonant_text += letter
            track_word += 1
            letter = word[track_word]

        intermediate = word[track_word:] + consonant_text

        if consonant_text[-1] + intermediate[0] == 'qu':
            intermediate = intermediate[1:] + 'u' + 'ay'
            intermediates.append(intermediate)
            continue

        else:
            intermediate = intermediate + 'ay'
            intermediates.append(intermediate)

    return ' '.join(intermediates)

    

```
