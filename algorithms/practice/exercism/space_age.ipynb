```python
class SpaceAge:
    """Calculate a person's age in years on each planet in the solar system.

    Ages are computed from a duration in seconds, using each planet's orbital
    period relative to Earth's (one Earth year = 31,557,600 seconds).

    Attributes:
        orbital_periods (dict): Orbital period of each planet relative to Earth.
        seconds (int | float): Age in seconds.
    """

    orbital_periods = {
        'Mercury': 0.2408467,
        'Venus': 0.61519726,
        'Earth': 1.0,
        'Mars': 1.8808158,
        'Jupiter': 11.862615,
        'Saturn': 29.447498,
        'Uranus': 84.016846,
        'Neptune': 164.79132
    }

    EARTH_YEAR_SECONDS = 31_557_600

    def __init__(self, seconds):
        """Initialize SpaceAge with a duration in seconds.

        Args:
            seconds (int | float): The age or duration to convert, in seconds.
        """
        self.seconds = seconds

    def age_on(self, planet):
        """Return age in years for the given planet.

        Args:
            planet (str): Planet name, must be a key in orbital_periods.

        Returns:
            float: Age in years on the specified planet, rounded to 2 decimal places.
        """
        return round(self.seconds / (self.EARTH_YEAR_SECONDS * self.orbital_periods[planet]), 2)

    def on_earth(self):
        """Return age in Earth years.

        Returns:
            float: Age in Earth years, rounded to 2 decimal places.
        """
        return self.age_on('Earth')

    def on_mercury(self):
        """Return age in Mercury years.

        Returns:
            float: Age in Mercury years, rounded to 2 decimal places.
        """
        return self.age_on('Mercury')

    def on_venus(self):
        """Return age in Venus years.

        Returns:
            float: Age in Venus years, rounded to 2 decimal places.
        """
        return self.age_on('Venus')

    def on_mars(self):
        """Return age in Mars years.

        Returns:
            float: Age in Mars years, rounded to 2 decimal places.
        """
        return self.age_on('Mars')

    def on_jupiter(self):
        """Return age in Jupiter years.

        Returns:
            float: Age in Jupiter years, rounded to 2 decimal places.
        """
        return self.age_on('Jupiter')

    def on_saturn(self):
        """Return age in Saturn years.

        Returns:
            float: Age in Saturn years, rounded to 2 decimal places.
        """
        return self.age_on('Saturn')

    def on_uranus(self):
        """Return age in Uranus years.

        Returns:
            float: Age in Uranus years, rounded to 2 decimal places.
        """
        return self.age_on('Uranus')

    def on_neptune(self):
        """Return age in Neptune years.

        Returns:
            float: Age in Neptune years, rounded to 2 decimal places.
        """
        return self.age_on('Neptune')
```
