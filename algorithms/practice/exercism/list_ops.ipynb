```python
def append(list1, list2):
    result = []
    result_pointer = 0
    if len(list1) > 0:
        result = [0]*len(list1)
        while result != list1:
            result[result_pointer] = list1[result_pointer]
            result_pointer += 1
    
    list_2_pointer = 0
    if len(list2) > 0:
        result = result + [0]*len(list2)
        while result[len(list1):] != list2:
            result[result_pointer] = list2[list_2_pointer]
            list_2_pointer += 1
            result_pointer += 1

    return result
```

```python
append([], [1,2,3,4])
```

```python
def concat(lists):
    result = []
    for list in lists:
        for item in list:
            result.append(item)

    return result
```
