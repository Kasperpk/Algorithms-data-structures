# Atbash Cipher

Create an implementation of the **Atbash cipher**, an ancient encryption system created in the Middle East.

## How it works

The Atbash cipher is a simple substitution cipher that transposes the alphabet so the resulting alphabet runs backwards. The first letter is replaced with the last, the second with the second-last, and so on.

For the Latin alphabet:

```text
Plain:  a b c d e f g h i j k l m n o p q r s t u v w x y z
Cipher: z y x w v u t s r q p o n m l k j i h g f e d c b a
```

It is a very weak cipher: it has only one possible key and is a simple mono-alphabetic substitution. This may not have been an issue in the cipher's time.

## Rules

- Ciphertext is written in **groups of 5 letters** (the traditional group size), separated by spaces, to hide word boundaries.
- **Numbers are left unchanged.**
- **Punctuation is excluded.**
- All text is encoded as **lowercase** letters.

````python
"""Module providing a functions for encoding and decoding atbash cipher text"""

import string

mapping_table = {'a': 'z', 'b': 'y', 'c': 'x', 'd': 'w', 'e': 'v',
                 'f': 'u', 'g': 't', 'h': 's', 'i': 'r', 'j': 'q',
                 'k': 'p', 'l': 'o', 'm': 'n', 'n': 'm', 'o': 'l',
                 'p': 'k', 'q': 'j', 'r': 'i', 's': 'h', 't': 'g',
                 'u': 'f', 'v': 'e', 'w': 'd', 'x': 'c', 'y': 'b',
                 'z': 'a'}
cipher_mapping = {key: value for (value, key) in mapping_table.items()}


def encode(plain_text):
    '''
    Function that encodes plain text into atbash cipher text
    '''
    
    res = ''
    for char in plain_text:
        if char.isalnum():
            res = res + mapping_table[char.lower()].lower()
        else:
            if char in string.punctuation:
                continue
            res = res + char
    res = res.replace(' ', '')
    return ' '.join(res[index:index+5] for index in range(0,len(res),5))

def decode(ciphered_text):
    '''
    Function that dencode atbash cipher text into plain text
    '''
    
    res = ''
    for char in ciphered_text:
        if char.isalpha():
            res = res + cipher_mapping[char.lower()].lower()
        else:
            if char in string.punctuation:
                continue
            res = res + char
    return res.replace(' ', '')
    
````

````python
"""Module providing a functions for encoding and decoding atbash cipher text"""

import string
from string import ascii_lowercase



mapping_table = {'a': 'z', 'b': 'y', 'c': 'x', 'd': 'w', 'e': 'v',
                 'f': 'u', 'g': 't', 'h': 's', 'i': 'r', 'j': 'q',
                 'k': 'p', 'l': 'o', 'm': 'n', 'n': 'm', 'o': 'l',
                 'p': 'k', 'q': 'j', 'r': 'i', 's': 'h', 't': 'g',
                 'u': 'f', 'v': 'e', 'w': 'd', 'x': 'c', 'y': 'b',
                 'z': 'a'}
cipher_mapping = {key: value for (value, key) in mapping_table.items()}

ENCODING = str.maketrans(ascii_lowercase, ascii_lowercase[::-1])

def encode(plain_text):
    '''
    Function that encodes plain text into atbash cipher text
    '''
    

    res = ''.join(char for char in plain_text.lower() if char.isalnum()).translate(ENCODING)
    return ' '.join(res[index:index+5] for index in range(0,len(res),5))

def decode(ciphered_text):
    '''
    Function that dencode atbash cipher text into plain text
    '''
    
    return "".join(char.lower() for char in ciphered_text if not char.isspace()).translate(ENCODING)
````
