In this problem we have to try and find the longest sequence of consequtive ones in a list of numbers. The first thought that came to my mind is that if we could split the array on the 0's then we could just return the maximum length of those sub arrays created. The split function only works on strings so i had to find a way to turn that list of numbers in to a string. To do that i used the `.join` function on a empty string for every number casted to a string. This allows one to concatenate string elements like what i wanted to do here. 

Then i could split the string on '0' and then return the length of the maximum value which will always correspond to the highest number like 111 is greater than 11 and so on. 

```python
def findMaxConsecutiveOnes(nums):
        string = ''.join([str(num) for num in nums])
        splits = string.split('0')

        return len(max(splits))        
```

```python
findMaxConsecutiveOnes([1,1,0,1,1,1])
```

There are like almost always other ways of solving this problem. A second approach uses two pointers and then iterate through the array. This is actually a faster algorithm because we don't have to cast every number to a string. This is an O(N) operation about as efficient as we can get for this type of problem.

```python
def findMaxConsecutiveOnes(nums):
    count = 0
    res = 0
    for num in nums:
        if num == 1:
            count += 1
            res = max(res, count)
        else: 
            count = 0
    return res 
```

```python
findMaxConsecutiveOnes([1,1,0,1,1,0,1,1,1])
```
