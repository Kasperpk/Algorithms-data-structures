You have a set of integers s, which originally contains all the numbers from 1 to n. Unfortunately, due to some error, one of the numbers in s got duplicated to another number in the set, which results in repetition of one number and loss of another number.

You are given an integer array nums representing the data status of this set after the error.

Find the number that occurs twice and the number that is missing and return them in the form of an array.

```python
def set_mismatch(nums):
    nums = sorted(nums)
    index = 0
    while index+1 < len(nums):
        if nums[index+1] == nums[index]:
            if nums[0] != 1:
                return [nums[index], 1]
            if index == 0 and nums[index] != 1:
                return [nums[index], 1]
            if nums[index] == 1:
                return [nums[index], 2]
            if nums[index] - nums[index-1] != 1:
                return [nums[index], nums[index]-1]
            return [nums[index], nums[index]+1]
        index += 1
```

```python
set_mismatch([2,3,2])
```

## Solution 2 — Sum + Dictionary

### Intuition
At first glance, looking at the example problems and solutions, I was only thinking that neighbouring numbers could be duplicates. So I created a loop that checked whether `index` and `index+1` were equal, and returned the duplicate and its successor. It turned out there were many edge cases I was not aware of, and I kept handling them one-by-one until I realised I needed a different approach.

### Approach
Instead of comparing neighbouring elements, I calculated the **actual sum** of the array and the **correct sum** using the Gauss formula $\frac{n(n+1)}{2}$.

I also created a dictionary initialised to `0` for every value from `1` to `n`, to count how many times each number appears.

The main loop walks the array from left to right:
1. Increment the counter for `nums[index]` in the dictionary.
2. If the counter exceeds `1`, we've found the duplicate.
3. The missing number is derived from the difference between the correct and actual sums:

$$\text{missing} = \text{correct\_sum} - \text{actual\_sum} + \text{duplicate}$$

**Example** — `[2, 3, 2]`:
- `actual_sum = 7`, `correct_sum = 6`
- At index 2 the counter for `2` reaches `2` → duplicate found.
- `missing = 6 - 7 + 2 = 1`
- Return `[2, 1]` ✓

### Complexity
- **Time:** $O(n)$ — single pass through the array.
- **Space:** $O(n)$ — the dictionary stores one entry per distinct value `1..n`.

```python
def set_mismatch(nums):
    actual_sum = sum(nums)
    n = len(nums)
    correct_sum = int(n*(n+1))/2
    dictionary = {key: 0 for key in range(1,len(nums)+1)}
    index = 0
    while index+1 <= len(nums):
        dictionary[nums[index]] = dictionary[nums[index]] + 1
        if dictionary[nums[index]] > 1:
            return [nums[index], int(correct_sum-actual_sum+nums[index])]
        index += 1
```

```python
set_mismatch([2,3,2])
```
