## Problem statement

You are responsible for collecting signatures from all tenants in a build
ing. For each tenant, you know a period of time when he or she is at home.
You would like to collect all signatures by visiting the building as few times
as possible. For simplicity, we assume that when you enter the building,
you instantly collect the signatures of all tenants that are in the building at that time.

### inputs

The first line of inputs is a number *n* which shows how many segments ther eare. Then there follows *n* lines which contain two integers separated by space defining the coordinates of endpoints of the *i*-th segment.

Examples
* 3 - number of segments
* 1 3 - first segment
* 2 5 - second segment
* 3 6 - third segment

We will then have a task like this where we have to find the minimum number of points we need to touch all segments. If they are all overlapping as here, then we can use a single point.




### outputs
The minimum number *k* of points on the first line and the integer coordinates of *k* points (separated by spaces) on the second line. You can output the points in any order. If there are multiple such sets of points, you can output any of them.

In the case above it would be
* 1 - number of points to connect all segments
* 3 - the point or coordinate 3 is the one that connects the dots

The safe choice here seems to be where the first points stops, so it seems we have to loop through the segments and put a point where the first segment stops and then remove all segments that lies within that coordinate. Then we start where the first segments next begins and that is the new starting point. 

```python
def collection_signatures(segments):
    num_stops = 0
    stops = []
    stop = segments[0][1]
    while segments:
        for segment in segments:
            # finding the first segment stop
            if segment[1] < stop:
                stop = segment[1]

        num_stops = num_stops + 1
        stops.append(stop)
        segments = [segment for segment in segments if segment[0] > stop]

    return num_stops, stops
```

```python
collection_signatures([[1,3], [2,5], [3,6]])
```
