# Introduction to Heaps

A **heap** is a specialized tree-based data structure used to quickly access the most important element.

## What a Heap Is

A heap is usually a **complete binary tree** (filled left-to-right), stored in an **array**.

Two common types:

- **Min-heap**: parent value <= children, so the smallest element is at the root.
- **Max-heap**: parent value >= children, so the largest element is at the root.

A heap is **not fully sorted**. It only guarantees parent-child ordering.

## Why Heaps Matter

Heaps are useful when you repeatedly need:

- smallest/largest element
- dynamic insertions and removals
- priority-based processing

Typical use cases:

- priority queues
- Dijkstra and A* (min-priority queue)
- top-k problems
- heap sort
- streaming median (using two heaps)

## Array Representation

For index `i`:

- left child: `2*i + 1`
- right child: `2*i + 2`
- parent: `(i - 1) // 2`

This makes heaps space-efficient and fast.

## Core Operations and Complexity

- peek min/max: `O(1)`
- insert: `O(log n)`
- extract min/max: `O(log n)`
- build heap from array (`heapify`): `O(n)`

## Build and Use Heaps in Python

Python's `heapq` module provides a **min-heap**.

```python
import heapq

h = []
heapq.heappush(h, 5)
heapq.heappush(h, 2)
heapq.heappush(h, 8)

print(h[0])              # 2
print(heapq.heappop(h))  # 2
print(heapq.heappop(h))  # 5
```

Build a heap from a list in `O(n)`:

```python
arr = [7, 1, 4, 9, 2]
heapq.heapify(arr)
```

Simulate a max-heap by pushing negative values:

```python
maxh = []
heapq.heappush(maxh, -10)
heapq.heappush(maxh, -3)
print(-heapq.heappop(maxh))  # 10
```

````python
def build_heap(array):
    start_index = len(array) // 2 - 1
    for index in range(start_index, -1, -1):
        left = index*2+1

        right = index*2+2

        while right < len(array):
            if array[index] < array[left] or array[index] < array[right]:
                max_node = max(array[left], array[right])
                max_node_index = array.index(max_node)
                array[index], array[max_node_index] = max_node, array[index]
                index = max_node_index
                left = max_node_index*2+1
                right = max_node_index*2+2
            else:
                break



    return array
````

````python
def max_heapify(array, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and array[left] > array[largest]:
        largest = left

    if right < n and array[right] > array[largest]:
        largest = right

    if largest != i:
        array[i], array[largest] = array[largest], array[i]
        max_heapify(array, n, largest)  # follow the swapped node down

def build_max_heap(array):
    n = len(array)
    for i in range(n // 2 - 1, -1, -1):  # -1 ensures root is included
        max_heapify(array, n, i)
    return array
````

````python
def heapsort(numbers):
    build_heap(numbers)
    sorted_numbers = []
    while numbers:
        sorted_numbers.append(numbers.pop(0))
        build_heap(numbers)
    return sorted_numbers
````
