# Dates and times in Python

Python's standard library handles dates and times through the **`datetime`** module. It is a good lens for object-oriented programming, because the module is really a small family of **classes**:

* `date` ‚Äî a calendar date (year, month, day).
* `time` ‚Äî a time of day (hour, minute, second, microsecond), independent of any date.
* `datetime` ‚Äî a single object that bundles **both** a `date` and a `time`.
* `timedelta` ‚Äî a *duration*, i.e. the difference between two points in time.
* `timezone` ‚Äî an offset from UTC.

Each of these is a **blueprint** (a class) from which you create **instances** (objects). Everything below is the same OOP vocabulary used in [classes.ipynb](classes.ipynb): the **constructor**, **class methods** and **class attributes** (which belong to the class itself), and **instance attributes** and **instance methods** (which belong to a particular object you created).

```python
import datetime
from datetime import date, time, timedelta, timezone
```

## The `datetime` class

A `datetime` object contains all the information from a `date` object **and** a `time` object. Like `date`, it assumes the current Gregorian calendar extended in both directions; like `time`, it assumes there are exactly `3600 * 24` seconds in every day.

### The constructor

You create an instance by **calling the class** (just like calling a function). The constructor takes `year`, `month`, `day`, and optionally `hour`, `minute`, `second`, and so on. Any time component you leave out defaults to `0`.

```python
d = datetime.datetime(2019, 1, 30)
d
```

### Class methods

Some methods are called **on the class itself** rather than on an instance, and they hand you back a brand-new object. `.now()` and `.today()` are alternative constructors of this kind ‚Äî both return the current moment as a `datetime`.

```python
datetime.datetime.now()
```

```python
datetime.datetime.today()
```

### Class attributes

The class also carries a few fixed values that describe the type as a whole (not any single object): `.min` and `.max` are the earliest and latest representable `datetime`, and `.resolution` is the smallest difference between two of them.

```python
print(datetime.datetime.min)
print(datetime.datetime.max)
print(datetime.datetime.resolution)
```

### Instance attributes

Each object you create stores its own data. These **instance attributes** are read directly off the instance `d` you built above.

```python
print(d.year)
print(d.month)
print(d.day)
print(d.hour)
print(d.second)
print(d.fold)
```

### Instance methods

**Instance methods** are functions you call on an object; Python passes the object in as `self` automatically. They let you derive other values from a `datetime` ‚Äî pull out just the date or time, ask which weekday it falls on, convert it to a timestamp, and so on.

```python
print(d.date())
print(d.time())
print(d.weekday())        # Monday is 0, Sunday is 6
print(d.isocalendar())    # (ISO year, ISO week, ISO weekday)
print(d.timetuple())
print(d.timestamp())
```

> **Instance vs. class.** A method like `weekday()` needs a real date to work on, so you call it on an instance: `d.weekday()`. Calling it on the bare class and passing an integer ‚Äî `datetime.datetime.weekday(5)` ‚Äî raises a `TypeError`, because `5` is not a `datetime` for `self` to bind to. The `()` after `d` is what hands the object in as `self`.

## The `date` class

When you only care about the calendar day (no time of day), use `date`. It mirrors `datetime`: the same idea of constructors, class methods, and instance methods.

### Alternative constructors

`date.today()` is a class method returning today's date, and `date.fromordinal()` builds a `date` from a *proleptic Gregorian ordinal* (day 1 is 0001-01-01). Both are ways to get a `date` instance without spelling out year/month/day yourself.

```python
date.today()
```

```python
d2 = date.fromordinal(734594)
d2
```

### Formatting a date

Instance methods turn a `date` into text. `.isoformat()` gives the standard `YYYY-MM-DD` string, `.strftime()` lets you choose your own layout with format directives, and `.ctime()` gives a fixed C-style string.

```python
d2.isoformat()
```

```python
print(d2.strftime("%d/%m/%y"))
print(d2.strftime("%A %d. %B %Y"))
```

```python
d2.ctime()
```

A few common `strftime` directives: `%Y` four-digit year, `%y` two-digit year, `%m` month number, `%B` month name, `%d` day of month, `%A` weekday name, `%H:%M:%S` 24-hour time.

## `timedelta` ‚Äî durations

Where `date` and `datetime` represent a *point* in time, **`timedelta`** represents a *span* of time ‚Äî for example "3 days" or "500 seconds". It is the bridge for **date arithmetic**:

* Subtracting two `datetime` (or two `date`) objects gives you a `timedelta`.
* Adding a `timedelta` to a `datetime` shifts it forward in time; subtracting one shifts it backward.

You construct one by naming the units you want, e.g. `timedelta(days=..., hours=..., seconds=...)`. That is the tool you reach for whenever you need "some amount of time after a given moment" ‚Äî which is exactly the shape of the gigasecond problem, left as your own exercise next door.

```python
delta = timedelta(days=30)
```

```python
print(timedelta.min)
print(timedelta.max)
print(timedelta.resolution)
```

```python
print(delta.total_seconds())
```
