```python
def job_scheduling(profits, time_available):
    sorted_jobs = sorted(profits, reverse=True)
    total_profits = 0
    for job in range(0, time_available):
        total_profits = total_profits + sorted_jobs[job]
    return total_profits
```

```python
job_scheduling([50,35,50,400,450,231,340], 5)
```

```python
def mice_and_fox(positions, holes):
    positions.sort()
    holes.sort()
    paired_array = []
    for position in range(0, len(positions)):
        paired_array.append((positions[position],holes[position]))
    return paired_array
```

```python
mice_and_fox([1,4,5,10,7], [3,2,4,5,8,9])
```
