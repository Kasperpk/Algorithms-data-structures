```python
def square(number):
    if number < 1 or number > 64:
        raise ValueError("square must be between 1 and 64")
    if number == 1:
        return 1

    return square(number-1) * 2
```

```python
square(3)
```

```python
def total(number):
    sum = 0
    for num in range(1,number+1):
        sum = sum + square(num)
    return sum
```

```python
total(4)
```
