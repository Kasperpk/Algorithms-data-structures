```python
fruits = ("apple", "banana", "cherry")
more_fruits = ["orange", "kiwi", "melon", "mango"]

combined_fruits = *fruits, *more_fruits
print(combined_fruits)

*combined_fruits, = *fruits, *more_fruits
print(combined_fruits)
```

```python
fruits_inventory = {"apple": 6, "banana": 2, "cherry": 3}
more_fruits_inventory = {"orange": 4, "kiwi": 1, "melon": 2, "mango": 3}

# fruits_inventory and more_fruits_inventory are unpacked into key-values pairs and combined.
combined_fruits_inventory = {**fruits_inventory, **more_fruits_inventory}

# then the pairs are packed into combined_fruits_inventory
combined_fruits_inventory
```

```python
# Arguments given to the function are packed into a tuple
def my_function(*args):
    print(args)

my_function(1, 2, 3)
my_function("Hello")

my_function(1, 2, 3, "Hello", "Mars")
```

```python
# Arguments given to the function are packed into a dictionary
def my_function(**kwargs):
    print(kwargs)

my_function(a=1, b=2, c=3)
```

```python
def my_function(*args, **kwargs):
    print(sum(args))
    for key, value in kwargs.items():
        print(str(key) + " = " + str(value))

my_function(1, 2, 3, a=1, b=2, c=3)
```

```python
def my_function(a, b, *args):
    print(a)
    print(b)
    print(*args)

my_function(1, 2, 3, 4, 5)
```

```python
def my_function(a, b, c):
    print(c)
    print(b)
    print(a)

numbers = [1, 2, 3]
my_function(*numbers)
```

I then use the * operator first to make a pack values into a combined list where if i left out the preceding * it would be a tuple. I reuse the * operator with every itereable i want to pack into this combined one. As you can see below i have several iterations. The first one i made the unpacking in several steps. In the second algorithm as my mentor suggested i put the unpacking into a single step, which is more clean

```python
def fix_list_of_wagons(each_wagons_id, missing_wagons):
    """Fix the list of wagons.

    :param each_wagons_id: list - the list of wagons.
    :param missing_wagons: list - the list of missing wagons.
    :return: list - list of wagons.
    """
    end_list = each_wagons_id[0], each_wagons_id[1]
    *wagons_updated, = each_wagons_id[2], *missing_wagons, *each_wagons_id[3:], *end_list
    return wagons_updated
```

```python
def fix_list_of_wagons(each_wagons_id, missing_wagons):
    """Fix the list of wagons.

    :param each_wagons_id: list - the list of wagons.
    :param missing_wagons: list - the list of missing wagons.
    :return: list - list of wagons.
    """
    first_wagon, second_wagon, locomotive, *remaining_wagons = each_wagons_id
    return [locomotive, *missing_wagons, *remaining_wagons, first_wagon, second_wagon]
```

```python
def fix_wagon_depot(wagons_rows):
    """Fix the list of rows of wagons.
    :param wagons_rows: list[list[tuple]] - the list of rows of wagons.
    :return: list[list[tuple]] - list of rows of wagons.
    """
    wagon_depots_zipped = zip(*wagons_rows)

    content = []
    for wagon_depot in wagon_depots_zipped:
        depot_row = []
        for depot in wagon_depot:
            depot_row.append(depot)
        content.append(depot_row)

    return content
```

```python
def fix_wagon_depot(wagons_rows):
    """Fix the list of rows of wagons.

    :param wagons_rows: list[list[tuple]] - the list of rows of wagons.
    :return: list[list[tuple]] - list of rows of wagons.
    """
    wagon_depots_zipped = zip(*wagons_rows)

    content = []
    for wagon_depot in wagon_depots_zipped:
        depot_row = [depot for depot in wagon_depot]
        content.append(depot_row)

    return content
```
