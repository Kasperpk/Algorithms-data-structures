```python
def depth_first_search(graph, start, visited=None, time=[0]):
    time[0] += 1
    if visited is None:
        visited = set()
    visited.add(start)
    for child in graph[start]:
        if child not in visited:
            depth_first_search(graph, child, visited, time)
    return time
```

```python
graph = {
      1: [2, 3],
      2: [4],
      3: [4],
      4: []
  }
```

```python
depth_first_search(graph, 3)
```
