### Problem statement

There are a total of numCourses courses you have to take, labeled from 0 to numCourses - 1. You are given an array prerequisites where prerequisites[i] = [ai, bi] indicates that you must take course bi first if you want to take course ai.

For example, the pair [0, 1], indicates that to take course 0 you have to first take course 1.
Return true if you can finish all courses. Otherwise, return false.

We can think of this problem as a directed graph, where the first element in each of the prerequisites pairs is what the second one depends on. In the example input given below we see that 1 depends on 0. We then can check if there is an element 0 that depends on 1. If that is the case then it's impossible to finish all the courses. So we have to check the second element and see if it depends on something we have visited, creating a cycle in the graph.

This problem is equivalent to finding if a cycle exists in a directed graph. If a cycle exists, no topological ordering exists and therefore it will be impossible to take all courses.


```python
def course_schedule(num_courses, prerequisites):
    visited = set()
    for prerequisite in prerequisites:
        print(f'to take the course {prerequisite[0]} you have to take course {prerequisite[1]}')
        print(f'therefore we add {prerequisite[1]} to the set')
        visited.add(prerequisite[1])
        for course in prerequisites:
            if course[0] in visited:
                return False
    return True

```

```python
from collections import defaultdict

def course_schedule(num_courses, prerequisites):
    adj = defaultdict(list)
    for course, prereq in prerequisites:
        adj[course].append(prereq)
    color = [0] * num_courses
    def dfs(u):
        color[u] = 1
        for v in adj[u]:
            if color[v] == 1:
                return False
            if color[v] == 0:
                if not dfs(v):
                    return False
        color[u] = 2
        return True
    for u in range(num_courses):
        if color[u] == 0:
            if not dfs(u):
                return False
    return True
```

```python
num_courses = 2
prerequisites = [[1,0], [0,1]]
```

```python
course_schedule(num_courses=num_courses, prerequisites=prerequisites)
```
