## Naive versus brilliant design

As i made my way through this problem i kept introducing new rules and i found it hard to keep the code concise and elegant as a result. Although i ended up passing the test, i had an idea that there would be major improvements that could be made to the code. I found in the community solutions a very brilliant design which only has 5 lines of code excluding the function definition and return statement. The code is clever because of the sequence that the code is executed in.

* It starts by replacing the '-' in the string
* It then checks if the length is different from 10, instead i have two checks which is unnecessary
* It then checks if the last digit is 'X' and if so replace it with 10.
* Then it uses a function `all` and checks if now not all are digits than it should return False because 'X' has been replaced now
* Finally it makes a very concise return statement summing the pairwise products of x,y zipping the nums and the decrementing range 10,0.

```python
def is_isbn(isbn):
    if '-' in isbn and len(isbn) != 13:
        return False
    if '-' not in isbn and len(isbn) != 10:
        return False
    character_check = [character for character in isbn if character.isalpha()]
    if len(character_check) == 1:
        if not isbn.endswith('X'):
            return False
    if len(character_check) > 1:
        return False
    isbn_groups = isbn.split('-')
    digits = ''.join(isbn_groups)
    sum = 0
    digit_index = 0
    for num in range(10,0,-1):
        if digits[digit_index] == 'X':
            sum = sum + num*10
            digit_index += 1
        else:
            sum = sum + num*int(digits[digit_index])
            digit_index += 1
    if sum%11 == 0:
        return True
    return False
```

```python
def is_isbn(isbn):
    nums = list(isbn.replace("-", ""))
    if len(nums)!=10:
        return False
    if nums[-1] == "X": nums[-1] = "10"
    if not all([c.isdigit() for c in nums]):
        return False
    return sum(int(x)*y for x,y in zip(nums, range(10, 0, -1)))%11 == 0
```
