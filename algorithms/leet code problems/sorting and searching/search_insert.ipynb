Given a sorted array of distinct integers and a target value, return the index if the target is found. If not, return the index where it would be if it were inserted in order.

You must write an algorithm with O(log n) runtime complexity.

## Naive (brute force) solution

The solution here below might have some challenges, but right now i am not sure how to overcome them. I wanted to use binary search but that repeatedly shrinks the array and so how do you return the index if not found.

There are several problems with the algorithm. For example if the distance between elements are above 99999 it will not update the closet element index. It also runs in linear time *O(n)* for which i know there are better solutions.

```python
class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        '''
        Function that returns the index of a target value in array
        of sorted numbers. If not found it returns the index which it
        should be placed in.
        :param nums: Array of sorted numbers
        :param target: Target value to find
        :return: Index of the position of target value in array
        '''

        # Tracking a closest element and index to return if
        # Target not found in array
        closest_element = 99999
        closest_element_index = 0

        # Using a for loop to find the target value or it's
        # Closest element index
        for index in range(0, len(nums)):

            # If the target - current index value in array is
            # less than current closest element then
            # This is a better position for it in the array
            # The difference has to be the closest to 0 and not
            # Large negative values.
            if abs(target - nums[index]) < closest_element:
                closest_element = target - nums[index]
                closest_element_index = index

            # if we find the target we just return the index
            if target == nums[index]:
                return index
        # Otherwise we return the index of the element closest to the
        # Target value
        else:
            if target >= nums[closest_element_index]:
                return closest_element_index + 1
            else:
                return closest_element_index
```

```python
result = Solution()
```

```python
result.searchInsert([-1,3,5,6], 0)
```

For a more optimal solution i had to talk to Claude about it, because i had tried my way with binary search but didn't know how to return the index of where the target should be placed if not found. It turns out there is quite a brilliant solution that defines a left and right variable in the initialization of the algorithm. Here `left` is 0 and `right` is the len of the array - 1. A while loop then proceeds until left no longer is smaller or equal to right. A variable `mid` is calculated and if nums[mid] is equal to target. Otherwise it checks if target is less than the mid point, if so `right` is changed to mid - 1, so the array is essentially halfed in either direction, the mid point is calculated again and the code proceeds in that way. This is quite a brilliant solution that i don't think i would have thought about, but using the left and right pointers (in general using pointers) just seems like a very crux move.

```python
class Solution:
    def search_insert(self, nums: List[int], target: int) -> int:
        left, right = 0, len(nums) - 1
        while left <= right:
            mid = (left + right) // 2
            if nums[mid] == target:
                return mid
            elif target < nums[mid]:
                    right = mid - 1
            else:
                left = mid + 1

        # When not found, left is the correct insert position
        return left
```

```python
result = Solution()
```

```python
result.search_insert([1,3,5,6], 7)
```
